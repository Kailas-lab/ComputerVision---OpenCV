```python
import cv2 as cv
import matplotlib.pyplot as plt

img=cv.imread(r"D:\kailas\open cv\ComputerVision---OpenCV\images\cats.jpg")
cv.imshow("cats",img)

gray=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('gray',gray)

cv.waitKey(0)
```

**Grayscale histogram**

cv.calcHist: This calculates the histogram of an image.
[gray]: The input image in grayscale.
[0]: Specifies the channel to calculate the histogram (0 indicates grayscale or the first channel).
None: No mask is applied; the histogram is calculated for the entire image.
[256]: The number of bins for the histogram (256 bins for grayscale intensity values 0-255).
[0, 256]: The range of intensity values to include in the histogram calculation.




```python
g_hist=cv.calcHist([gray], [0], None,[256],[0,256])

plt.figure()
plt.title("Grayscale histogram")
plt.xlabel("no. of pixels")
plt.ylabel("bins")

plt.plot(g_hist)

plt.xlim([0,256])
plt.show()

cv.waitKey(0)
```

**Color Histogram**

```python




# the image into B, G, R channels

colors = ('b', 'g', 'r')  # Corresponds to Blue, Green, and Red
plt.figure()
plt.title("Color Histogram")
plt.xlabel("Bins")
plt.ylabel("Number of Pixels")

# Calculate and plot the histogram for each channel
for i, color in enumerate(colors):
    hist = cv.calcHist([img], [i], None, [256], [0, 256])  # Compute histogram
    plt.plot(hist, color=color)  # Plot histogram with respective color
    plt.xlim([0, 256])

plt.show()
```

for i, color in enumerate(colors):

Loops through the indices and corresponding colors in the colors list.
i: Index of the channel (0 for blue, 1 for green, 2 for red).
color: The respective color string ('b', 'g', 'r') used for plotting.


hist = cv.calcHist([img], [i], None, [256], [0, 256])

Computes the histogram for the i-th channel of the image.
[img]: The input image (in BGR format).
[i]: Specifies which channel's histogram to compute (0 for Blue, 1 for Green, 2 for Red).
None: No mask is applied, meaning the histogram is calculated for the entire image.
[256]: Specifies 256 bins for the histogram (intensity values range from 0 to 255).
[0, 256]: The intensity range for the histogram.
